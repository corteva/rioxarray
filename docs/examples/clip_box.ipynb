```python
import rioxarray # for the extension to load
import xarray

%matplotlib inline
```

## Load in xarray dataset

```python
xds = xarray.open_dataarray("../../test/test_data/input/MODIS_ARRAY.nc")
```

```python
xds
```

```python
xds.plot()
```

## Clip using a bounding box

See docs for `rio.clip_box`:

  - [DataArray.clip_box](../rioxarray.rst#rioxarray.rioxarray.RasterArray.clip_box)
  - [Dataset.clip_box](../rioxarray.rst#rioxarray.rioxarray.RasterDataset.clip_box)

```python
xdsc = xds.rio.clip_box(
    minx=-7272967.1958741,
    miny=5048602.84382404,
    maxx=-7272503.88315758,
    maxy=5049066.15654056,
)
```

```python
xdsc
```

```python
xdsc.plot()
```

```python
xdsc = xds.rio.clip_box(
    minx=-7.3e+06,
    miny=4.99e+06,
    maxx=-7.2e+06,
    maxy=5.06e+06,
    allow_padding=True
)
```

```python
xdsc.plot()
```
