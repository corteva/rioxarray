## Pad using a bounding box

See docs for `rio.pad_box`:

  - [DataArray.pad_box](../rioxarray.rst#rioxarray.raster_array.RasterArray.pad_box)
  - [Dataset.pad_box](../rioxarray.rst#rioxarray.raster_dataset.RasterDataset.pad_box)

```python
xdsc = xds.rio.pad_box(
    minx=-7.3e+06,
    miny=4.99e+06,
    maxx=-7.2e+06,
    maxy=5.06e+06,
)
```

```python
xdsc.values[0, 0]
```

```python
xdsc.plot()
```
