# Example - Cloud Optimized GeoTiff (COG)

```python
import rioxarray

%matplotlib inline
```

```python
cog_url = (
    "https://esip-pangeo-uswest2.s3-us-west-2.amazonaws.com/"
    "sciencebase/Southern_California_Topobathy_DEM_1m_cog.tif"
)
```

```python
rds = rioxarray.open_rasterio(cog_url, overview_level=7)
```

```python
rds
```

```python
rds.where(rds!=rds.rio.nodata).plot()
```
